```python
d={'Delhi Daredevils':'Delhi Capitals','Kings XI Punjab':'Punjab Kings','Rising Pune Supergiants':'Pune Warriors','Rising Pune Supergiant':'Pune Warriors','Gujarat Titans':'Gujarat Lions'}
df=df.replace(d.keys(),d.values())
```

```python
df.nunique()
```

```python
df.Team1.unique()
```

```python
df.columns
```

```python
teams=df.Team1.unique()
l=[]
for i in teams:
    played=df[(df.Team1==i)|(df.Team2==i)].shape[0]
    win=df[df.WinningTeam==i].shape[0]
    home_win=df[(df.WinningTeam==i) & (df.Team1==i)].shape[0]
    home_play=df[df.Team1==i].shape[0]
    away_win=df[(df.WinningTeam==i) & (df.Team2==i)].shape[0]
    away_play=df[df.Team2==i].shape[0]
    l.append([i,played,win/played*100,home_win/home_play*100,away_win/away_play*100]) 
new=pd.DataFrame(l,columns=['team_name','home_win','home_play','away_win','away_play'])
print(new)
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('diabetes_unclean.csv')
df.head(7)
```

```python
df['Cr']=df.Cr.fillna(df.Cr.median())
df['HDL']=df.HDL.fillna(df.HDL.mode()[0])
df['AGE']=df.AGE.fillna(df.AGE.mean())
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df['Gender']=df['Gender'].replace('f','F')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(df['Cr'])
plt.show()
```

```python
#remove the outliers in Cr column
q1=df['Cr'].quantile(0.25)
q3=df['Cr'].quantile(0.75)
iqr=q3-q1
ul=q3+1.5*iqr
ll=q1-1.5*iqr
df=df[(df['Cr']>=ll) & (df['Cr']<=ul)]
plt.boxplot(df.Cr)
#plt.boxplot(df['Cr']) 
plt.show()
```
